パーセプトロン
パーセプトロン とは、脳の神経細胞にヒントを得て考案された、機械学習のモデルの一種です。パーセプトロンは、複数の入力を受け取り、1つの値を出力します。

ここではパーセプトロンの単純な例として、 petal_length と petal_width を入力値として使い、花の種類がSetosaか、それ以外か の分類結果を出力するモデルを作成してみましょう。

二値分類
2種類にグループ分けする分類問題を 二値分類 や 二項分類、 2クラス分類 と呼びます。このレッスンでは 二値分類 に統一します。対して3種類以上にグループ分けする分類問題は 多クラス分類 や 多項分類 と呼びます。

目的変数：花の種類がSetosaか、それ以外か
説明変数：petal_length, petal_width

```python
#二値分類の準備,まずはdatasetに花の種類がSetosaかそれ以外かの列を追加します。
# 「setosaかそれ以外か」の列
is_setosa = (dataset['species'] == 0).astype(np.int64)

# datasetに追加
dataset['is_setosa'] = is_setosa
dataset.head()
```

次に、petal_length, petal_width でsetosaを分類できるか、グラフにして確認してみましょう。seabornの lmplot を使うと、色分けした散布図を作成できます

パーセプトロンとは
パーセプトロンは、まず、受け取った⼊⼒ x1,x2,x3...に重みw1,w2,w3...を掛けて合計します。この合計から閾値（しきいち）と呼ばれる値θを引いたもの（ここではsとします）が0より⼤きければ1を出⼒し、そうでなければ0を出⼒します

パーセプトロンの学習則
パーセプトロンをアヤメの分類問題に適⽤するうえで、どのような流れでモデルが学習するか（学習則）を考えます。

ここでは、2つの説明変数 petal_width，petal_length の値をそれぞれx1,x2へ⼊⼒するとします。そして、⼊⼒した petal_width，petal_length の値を持つアヤメがsetosaならば1を出⼒し、それ以外ならば0を出⼒するように、重みw1,w2とw0を学習によって決めるものとします。

setosaだと判断した場合
まず、w0,w1,w2に適当な値を与えます。そして、学習データセットの中にある1つのアヤメの petal_width，petal_length の値（x1,x2）を⼊⼒します。

とある入力において、パーセプトロンの出⼒が1だった（setosaだと判断した）とします。この判定が正しければ、重みの値を変化させる必要はありません。しかし、この判断が誤りだった（実際はsetosa以外だった）場合には、出⼒が0になるように、それぞれの重みの値（ここではまとめてwiと表します）を変化させる必要があります。

ではwiをどのように変化させればよいのでしょうか。

0を出⼒させるためにはsが0以下になるように、sの中のw1x1を減らす必要があります。今考えているアヤメの分類問題の場合、長さの値であるx1（petal_width，petal_length）は正の値しか取らないことを考慮すれば、重みw1を減らせばよい、ということになります。

setosa以外だと判断した場合
今度は、パーセプトロンの出⼒が0だった（setosa以外だと判断した）とします。そしてこの判定が正しかった場合は、やはり重みwiを変化させる必要はありません。

しかし、この判定が誤りだった（実際はsetosaだった）場合には、出⼒が1になるように重みを変化させる必要があります。0だった出⼒を1にするためには、0以下だったs を増やして正の値となるようにしますが、そのためには先ほどと同様wiの値を増やせばよいわけです。

重みの更新に関するまとめ
以上の内容をまとめると、以下のようになります。

モデルの出力	実際の正解	重みの更新
0	             0	     必要なし
0	             1	     増やす
1	             0	     減らす
1	             1	     必要なし

パーセプトロンによる分類モデルを作成

```python
# 必要なライブラリの追加読み込み
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
```

```python
#今回は petal_width と petal_length を使って is_setosa を予測
# 目的変数（Y）：is_setosa、説明変数（X）：petal_widthとpetal_length
Y = np.array(dataset['is_setosa'])
X = np.array(dataset[['petal_width', 'petal_length']])
```

```python
# 形状を確認
print("Y=", Y.shape, ", X=", X.shape)
```

```python
# データの分割,ホールドアウト法
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
X_train, X_valid, Y_train, Y_valid = train_test_split(X_train, Y_train, test_size=0.3, random_state=0)
```

パーセプトロンは scikit-learn の Perceptron クラスを使います

```python
# パーセプトロンのモデルを作成し、学習して予測
perceptron_model = Perceptron()
perceptron_model.fit(X_train, Y_train)
Y_pred = perceptron_model.predict(X_valid)
```

```python
# 正解（検証データの目的変数）と予測値との比較
print(Y_valid[:5])
print(Y_pred[:5])
```

モデルの確認
重みがどうなったか確認してみましょう。パーセプトロンは線形なので、重みは線形回帰モデルと同様にして取得できます

```python
# 重みの確認
w0 = perceptron_model.intercept_[0]
w1 = perceptron_model.coef_[0][0] #coefは係数の意味
w2 = perceptron_model.coef_[0][1]

print("w0:", w0, ", w1:", w1, ", w2:", w2)
```

intercept_[0] にw0が、coef_[0][0] と coef_[0][1]に、それぞれ w1とw2が格納されています。（補足：この実行例ではw2は-0.7 に近い値となっていますが、乱数の初期値などにより実行ごとに値は変動する可能性があります。以下の説明では、この実行結果の例に基づき w2〜〜0.7として進めます。）
petal_width と petal_length の散布図に、w0+w1x1+w2x2=0で表される直線を描いてみましょう。この直線の方程式の左辺は、先ほど出てきたsと同じであることに注意してください（つまりs=0が直線を表す式になっています）。グラフへ直線を描くために、x2をx1の関数として変形します
では、上述の直線を petal_length, petal_width のグラフと重ねてみます（Seabornでグラフを重ね合わせるのは手間がかかるので、ここではmatplotlibを使っています）。

```python
# モデルの式
x1 = np.linspace(0, 3, 10)
x2 = (2 - 1.2 * x1) / 0.7

# 実際の値（setosa以外）
data = dataset.query("is_setosa == 0")
plt.scatter(data['petal_width'], data['petal_length'], color="blue", label="0")

# 実際の値（setosa)
data = dataset.query("is_setosa == 1")
plt.scatter(data['petal_width'], data['petal_length'], color="orange", label="1")

# モデルの式（setosaの部分）
plt.plot(x1, x2, color="green", linewidth=2)
plt.fill_between(x1, x2, color='green', linewidth=0, alpha=.05)

# グラフ全体の大きさを指定
plt.xlim(0, 2.8)
plt.ylim(0, 7.5)

# 凡例の表示
plt.legend(loc="upper left", title="is_setosa")
```

この図を⾒ると、パーセプトロンで求めた重みが係数と切⽚になる直線は、setosaとそれ以外を分ける直線であることがわかります
sのx1,x2に直線より上に位置する点の座標の値を⼊れると、負になることが確かめられます。たとえば、 (0.5, 3) の点で試してみましょう
s=2-1.2*0.5-0.7*3=-0.7
この場合、パーセプトロンの出⼒は0となり、petal_width，petal_lengthが (0.5, 3) の値を取るアヤメはsetosaではないと判断されたことになります
同様に、直線の下に位置する点の座標を⼊れると正になることが確かめられます。(0.2, 2) で試してみましょう。
s=2-1.2*0.2-0.7*2=0.36
この場合はパーセプトロンの出⼒は1となり、petal_width，petal_lengthが (0.2, 2) の値を取るアヤメはsetosaと判断されたことになります

モデルの評価
回帰モデルの評価は MSE(平均二乗誤差) を使って行いました。分類モデルの場合「どのぐらい正解しているか」が評価基準となります。観点によりいくつかの評価方法があります。このなかでは f1値 がもっとも多く使われます。
正解率（Accuracy）
適合率（Precision）
再現率（Recall）
f1値

正解率は「全体のなかで正解したものの割合」
正解率は scikit-learn の accuracy_score で求められます
モデルの score() を使用しても同様に求められる

```python
# 正解率
from sklearn.metrics import accuracy_score

accuracy_score(Y_valid, Y_pred)
```

```python
# モデルのscore関数による正解率
perceptron_model.score(X_valid, Y_valid)
```

適合率
適合率は「正解と判断した中で真に正解の割合」です。今回の場合、「setosaと判断したサンプルのなかで、実際のデータもsetosaで合っていた割合」が適合率
適合率は scikit-learn の precision_score で求められます

```python
# 適合率
from sklearn.metrics import precision_score

precision_score(Y_valid, Y_pred)
```

再現率は「真に正解のうち正解と判断した割合」です。今回の場合、「実際のデータがsetosaのサンプルのなかで、モデルもsetosaと判断した割合
再現率は scikit-learn の recall_score で求められます

```python
# 再現率
from sklearn.metrics import recall_score

recall_score(Y_valid, Y_pred)
```

F1値
F1値（F値）は、適合率と再現率の調和平均（逆数の平均を求め、それを逆数にした値）です。
F1値は scikit-learn の f1_score で求められます。

```python
# F1値
from sklearn.metrics import f1_score

f1_score(Y_valid, Y_pred)
```

まとめて出力
scikit-learn の classification_report を使うと、上記の値をまとめて出力できます。
recision：適合率、recall：再現率、f1:適合率と再現率の調和平均

```python
# モデルの評価をまとめて出力
from sklearn.metrics import classification_report

print(classification_report(Y_valid, Y_pred))
```

今回の実行結果は、すべて正しく分類ができたため、どの要素も 1.0 となっています。

ロジスティック回帰
パーセプトロンで「1か0」を分類できるようになりました。しかしこのままでは多クラス分類に対応できません。そこで機械学習のモデルの結果を「1か0」ではなく「setosaである確率」のように確率で出力する方法を考えてみましょう

シグモイド関数
機械学習のモデルの結果を 確率として表すために、パーセプトロンの考え方に以下のような変更を加えます。
機械学習のモデルの出力の値を「目的変数が特定の分類である確率」とする
出力の値は「0か1」ではなく連続的な確率となるようにする

シグモイド関数をパーセプトロンのモデルに適用

ロジスティック回帰の目的関数
線形回帰モデルの目的関数は誤差の2乗の合計を最小化するものでした。
ロジスティック回帰の出力は確率なので、目的関数も 「分類が正解となる確率」を最大化する ものとします。すべての学習データの事象が独立していると考えると、目的関数の数式は以下のようになります。これを 尤度関数 と呼びます。尤度とは「もっともらしさ（尤もらしさ）、どのくらい理にかなっているか」という意味です。Lは尤度の英語表記「Likelihood」の頭文字です。
尤度関数は確率の掛け合わせなので、値がものすごく小さくなります。そこで実際に計算する際は対数を取った 対数尤度関数 が良く使われています。

対数尤度と交差エントロピー
対数尤度関数は「分類が正解となる確率」を最大化するものなので、最適値も最大値を求めることとなります。同様の指標で 交差エントロピー（crossentropy、交差エントロピー誤差）というものもあります。交差エントロピーとは2つの確率分布の距離を表したものです。目的変数と予測値との距離のようなものになります。

もっと簡単に言うと、実は、交差エントロピーは対数尤度関数にマイナスを掛けたものになります。対数尤度を最大化することは、交差エントロピーを最小化することと同じです。交差エントロピーを用いるのは、最適化問題として最小化で考えたほうが解きやすい場合です。

具体的な計算式までは理解しなくても問題ありませんので 交差エントロピー という用語を覚えておくと良いでしょう。

ロジスティック回帰では多クラス分類を行うことができます。モデルの出力は 各クラスに属する確率 となるように拡張されています。このように「各クラスに属する確率」を表現できるように拡張した関数を ソフトマックス関数（softmax）と呼びます。ソフトマックス関数は、ディープラーニングの分類モデルなど、機械学習で多く用いられている関数の1つです。

それでは多クラス分類を行ってみましょう。ここではすべての説明変数を使用してアヤメの品種（species）を予測してみます。

```python
# Y:目的変数、X:説明変数
Y = np.array(dataset['species'])
X = np.array(dataset[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
# データの分割
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
X_train, X_valid, Y_train, Y_valid = train_test_split(X_train, Y_train, test_size=0.3, random_state=0)
```

クラスが 0, 1, 2 に増えました。これは、3つのクラスに分類されているから

決定木
決定木 (decision tree)とはデータを分割して分類を判定するモデル
決定木の途中の丸を ノード と呼びます。末端のノードは リーフ と呼ばれます。ノードとノードを繋ぐ線のことを リンク や エッジ と呼びます。決定木は判断の課程がわかりやすいため、機械学習だけでなく意思決定やデータの可視化の手法として広く用いられています。
決定木は、データのまとまりがなるべく単一のカテゴリーになるように、データを分割します。データに他のカテゴリーの値が入っている割合を 不純度 と呼びます。この不純度がなるべく小さくなるようにデータを分割します
不純度の計算にはジニ係数（ジニ不純度）や交差エントロピーといった計算式が使われます

```python
# 必要なライブラリの追加読み込み
from sklearn.tree import DecisionTreeClassifier
```

決定木のモデルを作成し、学習と予測、評価を行います。下記のコードでは、データを分割する基準（不純度）の計算式に交差エントロピーを指定しています（criterion='entropy'）。

```python
# 決定木のモデルを作成し、学習して予測
tree_model = DecisionTreeClassifier(criterion='entropy')
tree_model.fit(X_train, Y_train)
Y_pred = tree_model.predict(X_valid)

# モデルの評価
print(classification_report(Y_valid, Y_pred))
```

これまでのモデルより少し性能が悪いようです。前のレッスンで学んだグリッドサーチを使ってパラメータをチューニングしてみましょう

```python
# グリッドサーチを使う
from sklearn.model_selection import GridSearchCV

# モデル:決定木
model = DecisionTreeClassifier()

# max_depth:木の深さ, min_samples_split:ノードを分割するために必要なデータの数
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [5, 10, 50],
    'min_samples_split': [2, 4, 10]
}

# グリッドサーチ
gscv = GridSearchCV(model, param_grid=params, cv=3)

# データの分割:機械学習用データを学習と検証に分けるのはクロスバリデーションで行ってくれる
Xg_train, Xg_test, Yg_train, Yg_test = train_test_split(X, Y, test_size=0.3, random_state=0)

gscv.fit(Xg_train, Yg_train)
```

```python
# 最適なパラメータ
gscv.best_params_
```

```python
# 最適なパラメータによる決定木のモデル作成と学習、予測
tree_model = DecisionTreeClassifier(criterion='gini', max_depth=5, min_samples_split=2)
tree_model.fit(X_train, Y_train)
Y_pred = tree_model.predict(X_valid)

# モデルの評価
print(classification_report(Y_valid, Y_pred))
```

ランダムフォレスト
ランダムフォレストとは、複数の決定木を組み合わせて予測性能を高くしたモデルです。

・複数の決定木モデルを用意する
・それぞれの決定木の学習データは、もとの学習データから重複を許して同じ数をランダムに抽出したものを使用する（決定木ごとに微妙に学習データを変えることで学習のバリエーションを増やす）
・予測には、各決定木の予測結果の「多数決」を用いる

アンサンブル学習
ランダムフォレストの考え方は「予測性能の低いモデル（決定木）を組み合わせて予測性能の高いモデルをつくる」というものです。このような方法を一般に、アンサンブル学習（ensemble learning）と呼びます。アンサンブルはフランス語で「（組み合わせた）全体」という意味です。合奏や合唱でも使われる言葉です。

代表的なアンサンブル学習の方法には、以下の3つがあります。

・バギング（bagging）：学習データ全体から重複を許してサンプリングを行い、多数の等価なモデルを 並列 に学習させて多数決を取る方法。過学習を防ぎつつ、正確性を上げるのに有効。 ランダムフォレスト はバギングに含まれる
・ブースティング（boosting）：多数の等価な単位モデルを 直列 につなげる方法。前のモデルの予測誤りの情報を後のモデルが受け取り、より精度の高い予測を行うようにする。ランダムフォレストに 勾配ブースティング (Gradient Boosting) という手法を組み合わせた XGBoost や LightGBM というモデルが知られている
・スタッキング（stacking）：前のモデルの予測結果を後のモデルの入力として使うのはブースティングと同様だが、多数の必ずしも等価でないモデルを組み合わせた手法。ブースティングのモデルをさらに組み合わせるなど、より複雑なモデルを構築できる

ランダムフォレストは scikit-learn の RandomForestClassifier クラスを使います

```python
# 必要なライブラリの追加読み込み
from sklearn.ensemble import RandomForestClassifier
```

```python
# ランダムフォレストのモデルを作成し、学習して予測
randomforest_model = RandomForestClassifier(n_estimators=100)
randomforest_model.fit(X_train, Y_train)
Y_pred = randomforest_model.predict(X_valid)

# モデルの評価
print(classification_report(Y_valid, Y_pred))
```

サポートベクターマシン（SVM）
比較的高速で高い性能を出せるのがサポートベクターマシン（SVM）です

SVM
まずは基本的な「線形で二値分類するモデル」を例にします。これは 線形SVM と呼ばれます。

SVMとは、マージンを最大化するようにデータを分割する境界線（この例では2次元のため直線、高次元では超平面と呼ばれます。図では赤い実線）を決定するモデルです。マージンとは図の「青い矢印」を指し、各データと分割する線（赤い実線）との距離を表します。ここで、分割線を決めるのに重要なのは 分割線から近いデータ です。この分割線から近いデータを サポートベクトル と呼びます。

サポートベクトルからのマージンが最大になるように分割線を決める、というのが基本的な考え方です。
SVMは分割境界を決定するためにサポートベクトルのみを考慮するため、特にデータが高次元であっても効率的に学習できる場合があり、計算量が比較的少ないという特徴につながることがあります。

非線形の分類への応用
線形SVMでは直線でデータを分割します。直線で分割できない 非線形の分類 へ対応するために、SVMではカーネル法（カーネルトリック）という手法で 学習データを変換してから モデルに学習させる、という方法を取ります。

多クラス分類への応用
複数のSVMを組み合わせることで多クラス分類モデルを実現できます。例として A, B, C の3つに分類したい場合は、以下のどちらかの方法となります。両者とも予測時にはそれらのモデルの多数決で分類を決めます。

すべての組み合わせの3つのモデルを作成
「A, B」,「A, C」,「B, C」
1つとそれ以外の組み合わせの3つのモデルを作成
「A, BC」,「B, AC」,「C, AB」
